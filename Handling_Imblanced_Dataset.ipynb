A dataset with an unequal calss distribution

```python
import numpy as np
import pandas as pd
```

```python
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
data = credit_card_data
```

```python
data.head()
```

```python
data['Class'].value_counts()
```

This is Highly Imbalancef Dataset
0 --> Legit Transactions
1 --> Fradulant Transactions

```python
#separating the legit and fradulent transactions
legit = data[data.Class == 0]
fraud = data[data.Class ==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under- Sampling
Building a sample dataset containing similar distribution of legit and fradulant transactions
Numbre of fraudulent transactions --> 492

```python
legit_sample = legit.sample(n = 492)
```

```python
print(legit_sample.shape)
```

Concatenate the Two Dataframes

```python
new_dataset = pd.concat([legit_sample,fraud],axis = 0)
```
