Seaborn:
Data Visualization Library

```python
#importing the library
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note: Seaborn has some built-in datasets

```python
#total bill vs tip dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
#visualise the data
sns.relplot(data=tips, x='total_bill', y='tip',col ='time', hue='smoker',size='size')
```

```python
#setting default theme for the plots
sns.set_theme()
#only need to load one time
```

```python
#visualise the data
sns.relplot(data=tips, x='total_bill', y='tip',col ='time', hue='smoker',size='size')
```

```python
#loading the iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width',hue ='species', data = iris)
```

```python
sns.scatterplot(x='sepal_length', y='petal_width', hue='species', data = iris)
```

```python
#loading the titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
 sns.countplot(x ='class', data = titanic)
```

```python
#house price dataset
from sklearn.datasets import fetch_california_housing
house_cal = fetch_california_housing()
house = pd.DataFrame(house_cal.data, columns = house_cal.feature_names)
house['Price'] = house_cal.target
```

```python
print(house_cal)
```

```python
house.head()
```

 Distribution chart

```python
sns.distplot(house['Price'])
```

 Correlation:
 Positive
 Negative

```python
correlation = house.corr()
```

```python
#heatmap
#constructing heat map
plt.figure(figsize =(10,10))
sns.heatmap(correlation,cbar =True, square =True, fmt ='.1f', annot =True, annot_kws ={'size':8},cmap ='Blues')
```
