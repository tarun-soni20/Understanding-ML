Matplotlib: Useful for making plots and charts

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Ploting the data

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding title at x-axis and y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel("sine value")
plt.title('sine wave')
plt.show()
```

```python
#ploting a Parabola
x = np.linspace(-10,10,20)
y  = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'g+')
plt.show()
```

```python
plt.plot(x,y,'r.')
plt.show()
```

```python
x = np.linspace(-10,10,100)
plt.plot(x,np.sin(x), 'r-')
plt.plot(x,np.cos(x), 'b--')
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['eng','hin','urdu','punj']
people = [4,2,1,5]
ax.bar(languages,people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
lang = ['a','b','c','d']
people = [10,2,3,5]
ax.pie(people,labels = lang, autopct = '%1.1f%%')
plt.show()
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color ='r')
ax.scatter(x,z,color ='g')
plt.show()
```

```python
#3d scatter plot
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20* np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z, cmap ='Blues')
plt.show()
```
