```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

Ploting a vector

```python
plt.quiver(0,0,2,4)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy',angles ='xy' , scale =1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy',angles ='xy' , scale =1, color ='blue') #or 'b'
plt.quiver(0,0,-2,-6, scale_units='xy',angles ='xy' , scale =1, color ='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Addition of two vectors

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,4,7])
sum = vector_1 + vector_2
print(sum)
```

```python
plt.quiver(0,0,4,5, scale_units='xy',angles ='xy' , scale =1, color ='blue') #or 'b'
plt.quiver(0,0,-2,-6, scale_units='xy',angles ='xy' , scale =1, color ='y')
plt.quiver(0,0,2,-1, scale_units='xy',angles ='xy' , scale =1, color ='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Subtraction of 2 vectors

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = np.asarray([0,0,3,-2])

difference = vector_1 - vector_2
print(difference)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy', scale=1, color='y')
plt.quiver(0,0,-1,5, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Multiplying a vector by a scalar

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = 2*vector_1

print(vector_2)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,4,6, scale_units='xy', angles='xy', scale=1, color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = -0.5*vector_1

print(vector_2)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,-1,-1.5, scale_units='xy', angles='xy', scale=1, color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

 Dot Product of two vectors

```python
x = np.array([4,5])

y = np.array([2,3])

dot_product = np.dot(x,y)
```

```python
print(dot_product)
```

Cross product of two vectors

```python
a= np.array([20,25,30])

b = np.array([100,150,50])

a_cross_b = np.cross(a,b)
```

```python
print(a_cross_b)
```

Projection of "a" vector on "v" vector

```python
a = np.array([3,4])
v = np.array([1,2])

magnitude_of_v = np.sqrt(np.sum(v**2))
proj_of_a_on_v = (np.dot(a,v)/magnitude_of_v**2)*v
print(proj_of_a_on_v)
```

```python
a = np.array([3,4,5])
v = np.array([1,2,3])

magnitude_of_v = np.sqrt(np.sum(v**2))
proj_of_a_on_v = (np.dot(a,v)/magnitude_of_v**2)*v
print(proj_of_a_on_v)
```
